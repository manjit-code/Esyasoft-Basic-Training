9. Declare future vector(X) and target variable(y)

```python
X = df_3
y = df_3['status_type']
```

10. Convert categorical variable into integers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X['status_type'] = le.fit_transform(X['status_type'])
y = le.transform(y)
```

```python
X.info()
```

```python
X.head()
```

### Feature Scaling

```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()

X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

12. K-means model with two clusters

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

13. K-mean cluster center

```python
kmeans.cluster_centers_
```

```python
# Low inertia -> better model
kmeans.inertia_
```

14. Check Qulaity of weak classification by the model

```python
labels = kmeans.labels_

# Check how many of the samples were correctly labeled
correct_labeles = sum(y == labels)

print("Result: %d out of %d samples were correctly labled." % (correct_labeles, y.size))
```

```python
print('Accuracy score:  {0:0.2f}'.format(correct_labeles/float(y.size)))
```

```python
from sklearn.cluster import KMeans
cs = []

# List to store all accuracy value
accuracy = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
    labels = kmeans.labels_

    # Check how many of the samples were correctly labeled
    correct_labeles = sum(y == labels)
    accuracy.append(correct_labeles/float(y.size))

plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Numbe rof clusters')
plt.ylabel('cs')
plt.show()
```
