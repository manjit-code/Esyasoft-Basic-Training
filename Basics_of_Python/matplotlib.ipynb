```python
import matplotlib.pyplot as plt
```

```python
# Defining all the co-ordinates
x = [0,2,4,6,8]
y = [0,4,16,36,64]
```

```python
fig, ax = plt.subplots() 

ax.plot(x, y, marker='o', label='Data Points')

ax.set_title("Basic Components of Matplotlib Figure")

# Set Lable for Axis
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.legend() #Marker

plt.show() #To display the visulization
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random as rd
```

```python
arr = np.array([])
# arr = np.array([10, 20, 30])
for i in range(100):
    arr = np.append(arr, rd.randint(0, 100))
```

```python
df = pd.DataFrame({"Numbers" : arr})
plt.bar(df.index, df["Numbers"])
plt.show()
```
